```python
my_df = pd.read_csv('/content/Population data.csv')
my_df
```

```python
# Reshape the DataFrame
df_reshaped = pd.melt(my_df, id_vars=['country', 'country_code', 'id'], var_name='year', value_name='population')

# Convert 'year' column values to integers
df_reshaped['country'] = df_reshaped['country'].astype(str)
df_reshaped['year'] = df_reshaped['year'].astype(int)

# Replace commas in the 'population' column and convert to integers
df_reshaped['population'] = df_reshaped['population'].replace(',', '', regex=True).astype(int)

df_reshaped
```

```python
# Save reshaped data to CSV
df_reshaped.to_csv('population-data-2013-2022.csv')
```

```python
df = pd.read_csv("/content/population-data-2013-2022.csv")
df
```

```python
# Subset dataframe by year
selected_year = 2022
df_selected_year = df_reshaped[df_reshaped.year == selected_year]
df_selected_year
```

```python
# Sort by year
df_selected_year_sorted = df_selected_year.sort_values(by="population", ascending=False)
df_selected_year_sorted
```

```python
# Calculate population difference between selected and previous year
def calculate_population_difference(input_df, input_year):
  selected_year_data = input_df[input_df['year'] == input_year].reset_index()
  previous_year_data = input_df[input_df['year'] == input_year - 1].reset_index()
  selected_year_data['population_difference'] = selected_year_data.population.sub(previous_year_data.population, fill_value=0)
  selected_year_data['population_difference_absolute'] = abs(selected_year_data['population_difference'])
  return pd.concat([selected_year_data.country, selected_year_data.id, selected_year_data.population, selected_year_data.population_difference, selected_year_data.population_difference_absolute], axis=1).sort_values(by="population_difference", ascending=False)

df_population_difference_sorted = calculate_population_difference(df_reshaped, selected_year)
df_population_difference_sorted
```

```python
# Filter states with population difference > 50000
df_greater_50000 = df_population_difference_sorted[df_population_difference_sorted.population_difference_absolute > 50000]
df_greater_50000
```

```python
# % of States with population difference > 50000
int((len(df_greater_50000)/df_population_difference_sorted.country.nunique())*100)
```

```python
import plotly.express as px
import pandas as pd

# Assuming you have a 'Country Code' column in your DataFrame, you can use it for mapping
fig = px.choropleth(df_selected_year,
                    locations='country_code',
                    color='population',
                    color_continuous_scale='blues',
                    projection='natural earth',
                    title='Population by Country',
                    labels={'population': 'Population'}
                   )

# Show the figure
fig.show()
```

```python
# Create choropleth map
choropleth = px.choropleth(df_selected_year,
                            locations='country_code',
                            color='population',
                            locationmode="ISO-3",
                            color_continuous_scale='blues',
                            range_color=(0, max(df_selected_year['population'])),
                            scope="world",
                            labels={'population': 'Population'}
                           )

# Update layout
choropleth.update_layout(
    template='plotly_dark',
    plot_bgcolor='rgba(0, 0, 0, 0)',
    paper_bgcolor='rgba(0, 0, 0, 0)',
    margin=dict(l=0, r=0, t=0, b=0),
    height=350
)

# Display the choropleth map
choropleth.show()
```

# Objective:
The overarching aim of our project is to construct a robust and intuitive dashboard that harnesses a comprehensive dataset spanning the population statistics of diverse countries from 2013 to 2022. The dashboard is meticulously designed to furnish users with an interactive interface, allowing them to delve into intricate population trends, make cross-country comparisons, and extract nuanced insights into demographic shifts during the specified time period.

# Analysis:
1. **Dynamic Heat Map Visualization:**
   - At the heart of the dashboard lies a dynamic heat map, offering users the flexibility to visualize and interpret population variations across countries for each year. The incorporation of a color slicer provides users with the liberty to choose a color scale, enhancing the interpretability and depth of the displayed data.

2. **Identification of Highest and Least Populated Countries:**
   - The dashboard enables users to select a specific year using the slicer, dynamically presenting both the highest and least populated countries for the chosen year. This comparative analysis facilitates the identification of demographic outliers and aids in discerning overarching trends.

3. **Insightful Bar Chart Representation:**
   - Complementing the heat map is a sophisticated bar chart that graphically represents the population of all countries. This visualization is instrumental in providing an at-a-glance overview of population distributions, facilitating the identification of countries with noteworthy demographic shifts.

# Inferences & Insights:
1. **Discerning Temporal Population Changes:**
   - The dynamic heat map serves as a powerful tool for discerning temporal changes in population, enabling users to identify growth or decline trends in different countries over the specified time frame. These patterns contribute to a deeper understanding of the multifaceted factors influencing demographic changes.

2. **Facilitating Yearly Comparisons:**
   - The incorporation of a slicer allows users to seamlessly select specific years, fostering yearly comparisons. This feature empowers users to pinpoint countries experiencing significant population growth or decline, thereby aiding in a nuanced exploration of global demographic dynamics.

3. **Identifying Demographic Outliers:**
   - The dynamic display of the highest and least populated countries for each selected year facilitates the identification of demographic outliers. This information is invaluable for policymakers, researchers, and businesses seeking to focus attention on regions with distinctive demographic characteristics.

4. **Holistic Global Demographic Overview:**
   - The bar chart provides a comprehensive snapshot of population distribution across all countries. Users can expeditiously identify countries with the largest and smallest populations, contributing to a holistic understanding of global demographic patterns and their implications.

# Conclusion:
In conclusion, our meticulously crafted dashboard transforms raw population data into a visually compelling and interactively rich platform. This tool empowers users across various domains – from academia to policymaking and business planning – to explore, analyze, and derive meaningful insights from the dynamic and complex tapestry of global population dynamics.
